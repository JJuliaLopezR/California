#### Contenido de la base de datos:

Los datos pertenecen a las casas encontradas en un distrito de California dado y algunas estadísticas resumidas sobre ellas basadas en los datos del censo de 1990. Los datos no se limpiaron, por lo que haremos algunos pasos de preprocesamiento. Las columnas son las siguientes:

* longitude (longitud)
* latitude (latitud)
* housing_median_age (edad promedio de las casas del distrito)
* total_rooms (total de cuartos en el distrito)
* total_bedrooms (total de dormitorios en el distrito)
* population (población del distrito)
* households (número de casas)
* median_income (ingreso promedio de los hogares en el distrito)
* **median_house_value (valor promedio de las casas en el distrito) (variable objetivo)**
* ocean_proximity (cercanía del distrito al océano)


Lo que trataremos de hacer es ajustar una funcion tal que al evaluarlas con las demas caracteristicas esta pueda "predecir el valor de la variable objetivo"

## Cargar la información.

```python
#Primero importaremos algunas bibliotecas básicas que nos ayudarán en la exploración y preprocesamiento de datos.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Ahora cargamos el CSV como un DataFrame para poder trabajar.
data=pd.read_csv("housing_data.csv")
data
```

Nótese en la variable **ocean_proximity** cómo esta no tiene un valor numérico. No podemos simplemente tomar esos datos en formato de texto y meterlos en nuestro modelo. Por lo tanto, debemos preprocesar esta variable.

## Exploración de la información y división de la información.

```python
#Primero utilizaremos '.info()' para ver si tenemos valores nulos.
data.info()
```

Observamos que la mayoría de las variables en el conjunto de datos tienen 20,640 registros, sin embargo, la variable "total_bedrooms" solo tiene 20,433 registros, lo que implica una falta de 207 datos en esta variable. Dado que el conjunto de datos cuenta con más de 20,640 muestras, estas 207 muestras representan aproximadamente el 1% del total. Por lo tanto, podemos eliminar los distritos que no tienen este dato sin perjudicar significativamente la integridad de la base de datos.

```python
data.dropna(inplace=True)
data.info()
```

Ahora ya podemos ver que todas las variables tienen la misma cantidad de datos.

Ahora vamos a dividir los datos, esto para tener dos conjuntos: uno que utilizaremos para entrenar nuestro modelo y otro para evaluar nuestro modelo. Esto es importante hacerlo, ya que la evaluación debe hacerse con datos no vistos por el modelo.

```python
#Importamos la biblioteca necesaria para esto.
from sklearn.model_selection import train_test_split

#En el eje x pondremos todas las variables con las que contamos, exceptuando la variable objetivo.
X = data.drop(['median_house_value'], axis=1)
#En el eje y colocamos la variable objetivo.
y = data['median_house_value'] 
```

```python
#Ahora, sí separamos la información. La información para evaluación será solo el 20%.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Pero ahora tenemos las variables X_train e y_test separadas, lo cual no es muy conveniente para encontrar correlaciones. Por tanto, será mejor juntarlas de nuevo en un DataFrame.

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
#Creamos un histograma para ver las distribuciones de las variables.
train_data.hist(figsize=(15,8))
```

```python
#También es necesario ver la correlación entre la variable objetivo y las otras variables.
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu") #cmap=colores del mapa yellow green and blue 
```

Salta a la vista que la variable que muestra la mayor correlación es el ingreso promedio de los hogares en el distrito. Por lo tanto, esta variable se perfila como una de las más importantes para realizar observaciones y análisis posteriores. Además, también podemos observar una correlación inversa entre la latitud y el valor de las casas en el distrito.

## Preprocesamiento

Ahora notemos que los histogramas que hicimos anteriormente están sesgados hacia la izquierda en algunos de ellos. Tomemos el logaritmo de estas variables para que el histograma se parezca más a una curva gaussiana.

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

Ahora aprovechemos la variable **ocean_proximity**. Suena razonable pensar que una casa más cercana al océano será, en promedio, más costosa que una casa que esté muy alejada de este, especialmente si vives en un lugar como California. Sería interesante ver qué tan relacionada está esta característica con el valor promedio de las viviendas en el distrito. Para utilizar esta variable, primero debemos convertirla a un valor numérico. Analicemos los posibles casos.

```python
train_data.ocean_proximity.value_counts()
```

Notemos que tenemos 5 casos. Utilizaremos variables dummy para asignar un valor 0 o 1 si se cumple la característica, y eliminaremos la variable "ocean_proximity".

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

Volvemos a verificar la correlación, ahora con nuestras nuevas variables.

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

Notemos ahora que la relación de la distancia de estar a menos de una hora del océano es una característica que eleva el precio de las casas en el distrito. Y no solo eso, sino que estar **INLAND** es lo contrario, las casas en **INLAND** cuestan en promedio menos.

Para tener una mejor idea de esto, podemos trazar la ubicación de estos distritos, ya que tenemos la altitud y la latitud de cada uno.

```python
plt.figure(figsize=(15,8))
sns.scatterplot(y="latitude", x="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

[![Logo de OpenAI](https://store.mapsofworld.com/image/cache/data/D%20Maps%202014/California-Latitude-Longitude-Map-750Px-900x700.jpg)](https://store.mapsofworld.com/digital-maps/us-state-maps-and-counties-1-2-3/california/california-latitude-longitude-map)


Notemos que este es un mapa con la silueta de California, donde abajo a la izquierda se encuentra el mar. Como podemos observar, efectivamente se ve que las zonas costeras más cercanas al océano Pacífico tienen precios más altos. También notamos que a medida que nos movemos hacia el norte, el precio de las viviendas disminuye en comparación con las casas ubicadas cerca de la frontera con México y cerca de la ciudad de Los Ángeles.

## Creacion de caracteristicas

Tenemos algunas características que por sí solas nos brindan cierta información, pero tal vez si las combinamos podrían brindarnos información más valiosa para determinar el precio de la vivienda. Uno de estos ejemplos sería la razón de dormitorios entre el número total de habitaciones. Otro dato que se puede crear es el número promedio de habitaciones por vivienda en el distrito.

```python
train_data['bedroom_ratio']= train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/ train_data['households']
```

Revisamos nuevamente la correlación.

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

Notemos que la razón entre dormitorios y habitaciones es una característica muy importante y fue creada a partir de otras características que no eran tan importantes por sí solas. La correlación entre dormitorios y el precio de la casa era solo del 0.06, y la correlación entre habitaciones y el precio de la casa tenía un valor de 0.17, que aunque es buena, logramos crear una nueva característica que tuviera una correlación inversa del 0.2. No solo eso, también vemos que la característica de habitaciones por casa tiene una correlación del 0.11. Lo sorprendente de esto es que fue creada a partir de características (número de casas y número de habitaciones) que por sí solas solo tenían una correlación de -0.0079 y -0.17.

Esto es importante, ya que a la hora de entrenar modelos, es conveniente eliminar aquellas características cuya correlación con la variable objetivo sea cercana a cero.

## Modelo de regresión lineal.

Vamos a crear un modelo de regresión lineal que nos ayude a predecir el precio promedio de las viviendas en los distritos.

```python
#Importamos la biblioteca necesaria para hacer una regresión lineal.
from sklearn.linear_model import LinearRegression

#Indicamos nuestro eje x y nuestro eje y, donde el eje x 
#representa las características y el eje y representa la variable objetivo.
X_train, y_train = train_data.drop(['median_house_value'],axis=1), train_data['median_house_value']

#Entrenamos el modelo.
reg =LinearRegression()
reg.fit(X_train, y_train)
```

Antes de probar nuestro modelo en el conjunto de evaluación, debemos aplicar todas las transformaciones que realizamos al conjunto de entrenamiento también al conjunto de evaluación.

```python
#Aplicamos todas las transformaciones que aplicamos al conjunto de entrenamiento también al conjunto de evaluación.
test_data=X_test.join(y_test)

test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio']= test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/ test_data['households']
```

```python
#Indicamos cuáles serán las características y cuál será la variable objetivo.
X_test, y_test = test_data.drop(['median_house_value'],axis=1), test_data['median_house_value']
```

Ahora sí, evaluemos el modelo.

```python
reg.score(X_test, y_test)
```

Obtuvimos una precisión del 67%, lo cual no es muy mala pero tampoco muy buena. Por este motivo, será mejor entrenar otro modelo.

## Modelo de bosque de regresión aleatoria.

```python
#Importamos la biblioteca necesaria para realizar la regresión de bosque aleatorio.
from sklearn.ensemble import RandomForestRegressor

#Entrenamos el modelo.
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

Ahora sí, evaluamos el modelo de bosque aleatorio.

```python
forest.score(X_test, y_test)
```

Vemos que este modelo nos da una precisión mucho más alta, del 80%, lo cual es una excelente precisión. Observamos que el modelo de bosques de regresión aleatoria obtuvo una precisión significativamente mayor (80%) en comparación con el modelo de regresión lineal (67%). Esto indica que el modelo de bosques de regresión aleatoria es más efectivo para predecir el valor de las viviendas en el distrito de California.
